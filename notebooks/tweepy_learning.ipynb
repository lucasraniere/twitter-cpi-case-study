# Testing and learning the Tweepy library

```python
import tweepy
import json
```

Twitter credentials

```python
consumer_key = "*****"
consumer_secret = "***************"
access_token = "*****************"
access_secret = "*********"
```

Twitter authentication

```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

# API object
api = tweepy.API(auth)
```

### Searching for a given query (words, @users, hashtags, etc)

```python
hashtag_tweets = tweepy.Cursor(api.search, q="#CPIdaCovid", tweet_mode="extended").items(10)
tweet_array = []

for tweet in hashtag_tweets:
    tweet_array.append(tweet)
    print(tweet._json['full_text'])
```

```python
tweet_array[1]._json
```

### Seting the date of the retrieved tweets

```python
time_tweet = tweepy.Cursor(api.search, q="#CPIdaCovid", since="2021-5-15", tweet_mode="extended").items(5)

for tweet in time_tweet:
    print(tweet._json['full_text'])
```

### Searching tweet of a particular account

```python
account_tweets = tweepy.Cursor(api.user_timeline, screen_name="nytimes", tweet_mode="extended").items(5)

for tweet in account_tweets:
    print(tweet._json['full_text'])
```

```python
all_content = []
all_tweets = tweepy.Cursor(api.user_timeline, screen_name="nytimes", tweet_mode="extended").pages()

for page in all_tweets:
    for tweet in page:
        all_content.append(tweet._json)
        
len(all_content)
```
