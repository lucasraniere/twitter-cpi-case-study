```python
import pandas as pd
```

```python
DATA = 'data/gephi/'
```

```python
complete_network_nodes = pd.read_csv(DATA+'complete_nodes.csv')
complete_network_nodes.shape
```

```python
complete_network_nodes.sample(10)
```

```python
top_10_comm = complete_network_nodes['modularity_class'].value_counts()[:10]
```

```python
total_nodes = complete_network_nodes.shape[0]

for idx, val in zip(top_10_comm.index, top_10_comm.values):
    print(f'Community {idx}: {val} ({(val/total_nodes)*100:.3f}%) members')
```
